## Model 1: Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

#### Dropping 'id' column

```python
# Load and Prepare Data  by dropping 'id' column
X = df.iloc[:,1:-1]
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Create and Train the Logistic Regression Model
model1 = LogisticRegression(max_iter=500)
model1.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model1.predict(X_test)
```

```python
#Evaluate the Logistic Regression model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

### Logistic Regression Classification Report

#### Model Performance Overview

We employed a Logistic Regression model to perform binary classification on the dataset. The dataset consists of 30 features, and the target variable 'Class' represents binary outcomes (0 or 1). The Logistic Regression model was trained on a subset of the data and evaluated on a separate test set.

#### Key Model Metrics

1. **Accuracy:**
   - The Logistic Regression model achieved a commendable accuracy of approximately 96.30%. This indicates a strong ability to correctly predict the class for the instances in the test set.

2. **Precision, Recall, and F1-Score:**
   - Precision, recall, and F1-score are balanced, with class 0 showing slightly higher precision and class 1 slightly higher recall. The model performs well for both classes, achieving an overall balance in precision and recall.

3. **Confusion Matrix:**
   - The confusion matrix illustrates the model's performance, with 1,701 instances of class 0 misclassified as class 1 and 4,619 instances of class 1 misclassified as class 0. The majority of predictions fall along the diagonal, indicating correct classifications.

4. **Macro and Weighted Averages:**
   - The macro and weighted averages for precision, recall, and F1-score are all approximately 96%, demonstrating a consistent and reliable performance across the two classes.

#### Conclusion

The Logistic Regression model has shown robust performance on the test set, achieving a high level of accuracy and demonstrating a balanced trade-off between precision and recall. The confusion matrix highlights the model's ability to correctly classify the majority of instances, with a modest number of misclassifications.

### Model 2: Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=44)
```

```python
#Create and Train the Random Forest Classification Model
model2 = RandomForestClassifier(n_estimators=100, random_state=44)
model2.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model2.predict(X_test)
```

```python
#Evaluate the Random Forest model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
feature_importances = model2.feature_importances_
feature_names = X.columns

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)
```
